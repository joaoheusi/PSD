```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys

spf = open('alo.pcm', 'rb')
pcmdata = spf.read()
```

```python
a = np.frombuffer(pcmdata,np.int16)
```

```python
x = np.arange(len(a))
```

```python
plt.figure( figsize=(20,6))

plt.plot(x/2863, a)
#plt.ylim([-200,400])
plt.show()
```

```python
new_y = a * 1.5
```

```python
plt.figure( figsize=(20,6))
plt.plot(x/2863, new_y)
plt.show()
```

```python
saida_file = open('saida_alo.pcm', 'wb')
scaled = np.int16(new_y)
```

```python
saida_file.write(scaled.tobytes())
```

```python
saida_file.close()
```
